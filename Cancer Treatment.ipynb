```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
data1= load_breast_cancer()
```

```python
data1['data']
```

```python
cancer=pd.DataFrame(np.c_[data1['data'],data1['target']],columns=np.append(data1['feature_names'],'target'))
```

```python
cancer
```

```python
sns.pairplot(cancer,hue='target',vars=['mean radius','mean area'])
```

```python
sns.countplot(cancer['target'],label='count')
```

```python
x=cancer.drop(['target'],axis=1)
```

```python
y=cancer['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
from sklearn.svm import SVC
svc_model= SVC(random_state=10)
svc_model.fit(x_train,y_train)
```

```python
y_pred=svc_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

# Improving Model(TYpe 1)

```python
min_train=x_train.min()
max_train=x_train.max()
range_train=(max_train-min_train)
```

```python
x_train_scaled=(x_train-min_train)/range_train
```

```python
x_train_scaled.head(10)
```

```python
min_test=x_test.min()
max_test=x_test.max()
range_test=(max_test-min_test)
x_test_scaled=(x_test-min_test)/range_test
```

```python
from sklearn.svm import SVC
svc_model= SVC(random_state=10)
svc_model.fit(x_train_scaled,y_train)
y_predict=svc_model.predict(x_test_scaled)
```

```python
cm1=confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm1,annot=True,fmt='d')
```

```python
print(classification_report(y_test,y_predict))
```

# Improving Model by Feature Scaling(Type 2)

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_trn=sc.fit_transform(x_train)
x_tst=sc.transform(x_test)
```

```python
from sklearn.svm import SVC
svc_model= SVC(random_state=10)
svc_model.fit(x_trn,y_train)
y_predct=svc_model.predict(x_tst)
```

```python
cm2=confusion_matrix(y_test,y_predct)
```

```python
sns.heatmap(cm2,annot=True)
```

```python
print(classification_report(y_test,y_predct))
```
